```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
from config import api_key
from citipy import citipy
import openweathermapy.core as owm
```

```python
output_file_path=os.path.join('output','cities.csv')
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1000)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = (citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name).title()
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))

if len(cities)<500:
    # Create a set of random lat and lng combinations
    lats = np.random.uniform(low=-90.000, high=90.000, size=1000)
    lngs = np.random.uniform(low=-180.000, high=180.000, size=1000)
    lat_lngs = zip(lats, lngs)

#     Identify nearest city for each lat, lng combination
    for lat_lng in lat_lngs:
        city = (citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name).title()

        # If the city is unique, then add it to a our cities list
        if city not in cities:
            cities.append(city)
print(len(cities))
```

```python
#settings for owm
settings={'units':'imperial','appid':api_key}
iteration=0
temp=[]
lat=[]
humid=[]
cloud=[]
wind=[]
#iterate through dataframe and update with relevant data
for city in cities:
    print(f"Retriving data for Index {iteration}: {city}.")
    try:
        current_weather = owm.get_current(city, **settings)
        temp.append(current_weather['main']['temp'])
        lat.append(current_weather['coord']['lat'])
        humid.append(current_weather['main']['humidity'])
        cloud.append(current_weather['clouds']['all'])
        wind.append(current_weather['wind']['speed'])
    except:
        print('Data missing. Skipping')
        temp.append(np.nan)
        lat.append(np.nan)
        humid.append(np.nan)
        cloud.append(np.nan)
        wind.append(np.nan)  
        
    print("----------")
    iteration=iteration+1
```

```python
#Ensure all lists are equal lengths
print(len(cities))
print(len(lat))
print(len(temp))
print(len(humid))
print(len(cloud))
print(len(wind))
```

```python
# create dataframe
cities_df=pd.DataFrame({'City':cities,
                       'Latitude':lat,
                        'Temperature (F)':temp,
                        'Humidity (%)':humid,
                        'Cloudiness (%)':cloud,
                        "Wind Speed (mph)":wind})
cities_df.head(30)
```

```python
cities_df=cities_df.dropna(how='any')
cities_df.head(30)
```

```python
#Export to CSV
cities_df.to_csv(output_file_path)
```

```python
#create Temp vs. Latitude Graph
plt.scatter(cities_df['Latitude'],cities_df['Temperature (F)'],c='blue',edgecolor='black')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.title('Current Temperature vs. Latitude (3/11/19)')
plt.xlim(lat_range)
plt.ylim(0,100)
plt.grid()
plt.show()
plt.savefig('temp_vs.lat.png')
```

```python
#Create Humidty vs. Latitude graph
plt.scatter(cities_df['Latitude'],cities_df['Humidity (%)'],c='yellow',edgecolor='black')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.title('Current Humidity(%) vs. Latitude (3/11/19)')
plt.xlim(lat_range)
plt.ylim(-5,105)
plt.grid()
plt.show()
plt.savefig('humid_vs.lat.png')
```

```python
#create cloudiness vs. latitude graph
plt.scatter(cities_df['Latitude'],cities_df['Cloudiness (%)'],c='lightblue',edgecolor='black')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.title('Current Cloudiness(%) vs. Latitude (3/11/19)')
plt.xlim(lat_range)
plt.ylim(-5,105)
plt.grid()
plt.show()
plt.savefig('cloud_vs.lat.png')
```

```python
#Create Wind Speed vs Lat. Graph
plt.scatter(cities_df['Latitude'],cities_df["Wind Speed (mph)"],c='purple',edgecolor='black')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (MPH)')
plt.title('Current Wind Speed vs. Latitude')
plt.xlim(lat_range)
plt.ylim(0,40)
plt.grid()
plt.show()
plt.savefig('wind_vs.lat.png')
```

<h1> Observable Trends </h1>

* Temperature is correlated with Latitude. Cities closer to the equator have a warmer temperature (as we except on a Caribbean vacation!) while cities farther away are cooler. However, cities above the equator are in general cooler than cities on the equal northern latitude. 

* Cities that are closer to the equator are overall more humid than cities that are farther away. However, otherwise, there doesn't seem to be huge variation in humidity by latitude. It was raining (or otherwise precipitating) in quite a lot of the sampled cities when this data was taken. Perhaps we are in a rainy season currently.

* Cloudiness and wind speed don't seem to correlate by latitude. However, there are few data points to correspond to the extremes between 40-80 latitude as there are very few land masses and thus cities around that latitude. Perhaps, the winds are similar but our data here doesn't accurately assess those points.
